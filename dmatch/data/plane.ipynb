```python
import torch
from plane import *
from nsf_utils import torchutils
from matplotlib import pyplot as plt
```

```python
from utils.io import get_top_dir
from utils.plotting import plot2Dhist
def _test(dataset, nm=None):
    # device = torch.device('cuda')
    # torch.set_default_tensor_type('torch.cuda.FloatTensor')

    data = torchutils.tensor2numpy(dataset.data)
    fig, ax = plt.subplots(1, 1, figsize=(5, 5))
    # ax.scatter(data[:, 0], data[:, 1], s=2, alpha=0.5)
    bound = 4
    bounds = [[-bound, bound], [-bound, bound]]
    # bounds = [
    #     [0, 1],
    #     [0, 1]
    # ]
    # ax.hist2d(data[:, 0], data[:, 1], bins=256, range=bounds)
    # ax.set_xlim(bounds[0])
    # ax.set_ylim(bounds[1])
    plot2Dhist(data, ax, 200)
    # SineWaveDataset
    sz = 14
    # Diamond
    # sz = 16
    plt.xticks([-3, 0, 3], fontsize=sz)
    plt.yticks([-3, 0, 3], fontsize=sz)
    ax.tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False, labelbottom=False,
               labelleft=False)
    fig.tight_layout()
    if nm:
        fig.savefig(get_top_dir() + '/images/' + nm + '.png')
    plt.show()
```

```python
dataset = GaussianDataset(num_points=int(1e6))
_test(dataset, 'gauss')
```

```python
dataset = DiamondDataset(num_points=int(1e6), width=20, bound=2.5, std=0.04)
_test(dataset, 'diamond_ticks')
```

```python
dataset = CrescentDataset(num_points=int(1e6))
_test(dataset)
```

```python
dataset = CrescentCubedDataset(num_points=int(1e6))
_test(dataset)
```

```python
dataset = SineWaveDataset(num_points=int(1e6))
_test(dataset, 'sine_wave_ticks')
```

```python
dataset = AbsDataset(num_points=int(1e6))
_test(dataset)
```

```python
dataset = TestGridDataset(num_points_per_axis=50, bounds = [[-4,4],[-4,4]])
_test(dataset)
```
